```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargamos el CSV que generó tu scraper
df = pd.read_csv('../data/ofertas_tecnoempleo.csv')

# Mostramos las primeras 5 filas para verificar
df.head()
```

```python
# 1. Limpiamos la empresa
df['empresa'] = df['empresa'].str.strip().str.upper()

# 2. Una limpieza rápida de la columna ciudad para quedarnos solo con lo que hay antes del guión
df['ciudad_limpia'] = df['ciudad'].str.split('-').str[0].str.strip()

# Mostramos
df[['empresa', 'ciudad_limpia']].head()
```

```python
# Contamos cuántas ofertas hay por empresa
top_empresas = df['empresa'].value_counts().head(10)

# Configuramos el gráfico
plt.figure(figsize=(12, 6))
top_empresas.plot(kind='barh', color='#3498db', edgecolor='black')

# Añadimos títulos y etiquetas
plt.title('Top 10 Empresas con más ofertas de Data Analyst', fontsize=16, fontweight='bold')
plt.xlabel('Número de ofertas', fontsize=12)
plt.ylabel('Empresa', fontsize=12)
plt.gca().invert_yaxis()  # Para que la empresa con más ofertas esté arriba
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Mostramos el resultado
plt.show()
```
